```python
import schemdraw
from schemdraw.flow import *
```

```python
with schemdraw.Drawing() as d:
    d.config(fontsize=13)
    d += Start(w=5).label("AMS Workflow Starts")
    d += Arrow().down(d.unit/2)
    
    d += Data(w = 5).label("Menu recipes given \nin PDF")
    d += Arrow().down(d.unit/2)
    
    d += Process(w = 5).label("Read PDF files with \nPDF Parsing")
    d += Arrow().down(d.unit/2)
    
    d += (act1 := Process(w = 5).label("Generate all \ningredients list"))
    d += Arrow().down(d.unit/2)
    
    d += Process(w = 5).label("Analyze \nemission impacts for \neach ingredient")
    d += Arrow().down(d.unit/2)
    
    d += Process(w = 5, h=3).label("Compile all ingredients \nand emission factors \nfor each product")
    d += Arrow().down(d.unit/2)
    
    d += (decision := Decision(w=6, h=5, S="True", E="False").
         label("Is the product \nindependent? \n(not a Child Product)"))
    d += Arrow().down(d.unit/2)
    
    d+= (end2 := Box(w=4, h=2).label("Product label \nassigned"))
    
    d+= Arrow().right(d.unit).at(decision.E)
    d+= (end := Process(w=5).label("Isolate ingredients \nfor the Child Product")).drop('E').fill("lightpink")
    
    d+= Wire("c", k=1, arrow="->").at(end.E).to(act1.E).color('deeppink').label('List of \ningredients for \nChild Products are \nseparated', fontsize=12)
    d+= Wire("c", k=-1, arrow="->").at(end2.W).to(act1.W).color('deeppink').label('Child Product \nadded to \ningredient \nlist', fontsize=12)
    
    d.save('image/AMS_workflow.png')
```
