```python
import pandas as pd
from pandas.util.testing import assert_frame_equal
import numpy as np
```

# Validating Results

```python
def equal_dataframes(df1, df2):
    try:
        assert_frame_equal(df1, df2)
        return print("They are equal!")
    except:
        return print("They are not equal.")
```

## Validating results on OK21-22

```python
Jenny_results = pd.read_csv("Jenny_results_OK21_22.csv")
Jenny_results = Jenny_results.drop(["GHG Only Label", "Combined Label"], axis=1)
Jenny_results
```

```python
Silvia_results = pd.read_csv("Silvia_results_OK21_22.csv")
Silvia_results
```

```python
equal_dataframes(Silvia_results, Jenny_results)
```

#### Deliberately change a value of single cell to confirm that the function is working properly.

```python
# Try changing a value in Jenny_results to confirm if the above function is working properly
Jenny_results.at[0, "Weight (g)"] = 180.0
Jenny_results
```

```python
equal_dataframes(Silvia_results, Jenny_results)
```

### Find out which two rows return inconsistent values

```python
# Finds out which rows are different between two dataframes. 
## Returns nothing if there is no difference between two dataframes. 

def diff_dataframes(df1, df2):
    diff_df = df1[(df2 != df1).any(axis=1)]
    return diff_df
```

```python
diff_dataframes(Jenny_results_new, Silvia_results_new)
```
