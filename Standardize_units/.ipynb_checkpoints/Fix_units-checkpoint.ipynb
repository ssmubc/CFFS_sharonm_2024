```python
import pandas as pd
import numpy as np
```

```python
totem = pd.read_csv('C:/Users/smvan/CFFS-S23/CFFS-22-23/data/cleaning/non_std_items_for_restaurants/Items_standarized_for_Totem.csv')
gather = pd.read_csv('C:/Users/smvan/CFFS-S23/CFFS-22-23/data/cleaning/non_std_items_for_restaurants/Items_Nonstd_for_Gather.csv')
ok = pd.read_csv('C:/Users/smvan/CFFS-S23/CFFS-22-23/data/cleaning/non_std_items_for_restaurants/Items_Nonstd_for_OK.csv')
```

```python
mapping = pd.read_csv('C:/Users/smvan/CFFS-S23/CFFS-22-23/data/mapping/Mapping.csv')
```

```python
for index, row in totem.iterrows():
    ing = row['IngredientId']
    cat_id = mapping[mapping['ItemId'] == ing]['CategoryID'].iat[0]
    ghg = mapping[mapping['ItemId'] == ing]['Active Total Supply Chain Emissions (kg CO2 / kg food)'].iat[0]
    invent = mapping[mapping['ItemId'] == ing]['InventoryGroup'].iat[0]
    totem.loc[index, 'InventoryGroup'] = invent
    totem.loc[index, 'CategoryID'] = cat_id
    totem.loc[index, 'GHG_factor'] = ghg
```
